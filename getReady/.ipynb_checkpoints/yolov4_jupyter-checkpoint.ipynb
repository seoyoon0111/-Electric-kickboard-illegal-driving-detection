```python
# -*- coding: utf-8 -*-
import cv2
import numpy as np
import time
import threading
```

```python
# 학습된 yolo weight파일과 cfg 파일 가져오기
net = cv2.dnn.readNet("yolov4-obj_1000.weights", "yolov4-obj.cfg") # Helmet : 0, No Helmet : 1
```

```python
# obj.names 
classes = []
with open("obj.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]
colors = np.random.uniform(0, 255, size=(len(classes), 3))
print(colors)
```

```python
# 멀티 쓰레드는 영상과 경고음이 충돌하는 것을 막음
class Worker(threading.Thread):
    def __init__(self):
        self.flag = 0
        super(Worker, self).__init__()
```

```python
# 경보음 음성파일 출력 함수
from IPython.display import Audio 
def sound():
    display(Audio('sounds_warning.wav', autoplay=True))
    
# sound()
```

# 실시간 웹캠 yolo

```python
# 멀티 쓰레드는 영상과 경고음이 충돌하는 것을 막음
class Worker(threading.Thread):
    def __init__(self):
        self.flag = 0
        super(Worker, self).__init__()

# Loading camera
cap = cv2.VideoCapture(0)    
# 캡쳐 프레임 사이즈 조절
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 416)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 416)

font = cv2.FONT_HERSHEY_PLAIN
starting_time = time.time()

frame_id = 0
wList = []
for i in range(10):
    w = Worker()
    wList.append(w)
ti=0
L = []
while True:
    _, frame = cap.read()
    frame_id += 1
    height, width, channels = frame.shape
    # Detecting objects
    #416*416 <=== 정확도 조절
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)
    # Showing informations on the screen
    class_ids = []
    confidences = []
    boxes = []

    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                # Object detected
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)   # width of object
                h = int(detection[3] * height)
                # Rectangle coordinates
                x = int(center_x - w / 2)       #the starting X position of detected object
                y = int(center_y - h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))   #percentage
                class_ids.append(class_id)#the name of detected object
        indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.3)
    
    for i in range(len(boxes)):
        if i in indexes:
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            confidence = confidences[i]
            color = colors[class_ids[i]]
            cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
            cv2.rectangle(frame, (x, y), (x + w, y + 30), color, -1)
            cv2.putText(frame, label + " " + str(round(confidence, 2)), (x, y + 30), font, 3, (255,255,255), 3)
    
            if class_ids[i] == 1:  # Helmet : 0, No_Helmet : 1
                L.append(class_ids[i])
                
    #frame_id는 몇 프레임 마다 한번씩 체크를 할껀지 정하는 변수
    # 영상으로 헬멧인식을 할 경우 1~2초 설정?
    # 만약 헬멧이 인식되면 프로그램이 끝나고 헬멧을 쓰지 않은 경우는 경고음으로?
    # L.count(1)은 프레임 마다 검사를 했을 때, No_Helmet이라고 판단하는 변수가 몇 이상일 때 경보음을 울릴 것인가 경계값을 정하는 변수 
    if(frame_id % 10==0 and L.count(1)>=3): # 우선 test용으로 10초에 한번씩 노헬멧이 3번이상 인식되면 경보음
        print("헬멧 미착용")
        sound()
#         w = Worker()
#         w.start()
#         ti+=1
#         L.clear()

    elapsed_time = time.time() - starting_time
    fps = frame_id / elapsed_time
    cv2.putText(frame, "FPS: " + str(round(fps, 2)), (10, 50), font, 3, (0, 0, 0), 3)
    cv2.imshow("Helmet Detection", frame)
    key = cv2.waitKey(1)
    if key == ord('q'):  # q 두번 누르면 창 닫힘
        break

# wList[ti].join()
cap.release()
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 핸드폰 카메라 연결 - 영상
# Loading camera
# url = "https://192.168.35.219:8080/video"
url = "https://172.30.107.32:8080/video"

cap = cv2.VideoCapture(url)
# cv2.namedWindow('Helmet Detection', cv2.WINDOW_AUTOSIZE) # 캡쳐 프레임 사이즈 조절
cv2.namedWindow('Helmet Detection', cv2.WINDOW_AUTOSIZE)

font = cv2.FONT_HERSHEY_PLAIN
starting_time = time.time()

frame_id = 0
wList = []
for i in range(10):
    w = Worker()
    wList.append(w)
ti=0
L = []
while True:
    _, frame = cap.read()
    frame_id += 1
    height, width, channels = frame.shape
    # Detecting objects
    #416*416 <=== 정확도 조절
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)
    # Showing informations on the screen
    class_ids = []
    confidences = []
    boxes = []

    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                # Object detected
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)   # width of object
                h = int(detection[3] * height)
                # Rectangle coordinates
                x = int(center_x - w / 2)       #the starting X position of detected object
                y = int(center_y - h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))   #percentage
                class_ids.append(class_id)#the name of detected object
        indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.3)
    
    for i in range(len(boxes)):
        if i in indexes:
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            confidence = confidences[i]
            color = colors[class_ids[i]]
            cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
            cv2.rectangle(frame, (x, y), (x + w, y + 30), color, -1)
            cv2.putText(frame, label + " " + str(round(confidence, 2)), (x, y + 30), font, 3, (255,255,255), 3)
    
            if class_ids[i] == 1:  # Helmet : 0, No_Helmet : 1
                L.append(class_ids[i])
                
    #frame_id는 몇 프레임 마다 한번씩 체크를 할껀지 정하는 변수
    # 영상으로 헬멧인식을 할 경우 1~2초 설정?
    # 만약 헬멧이 인식되면 프로그램이 끝나고 헬멧을 쓰지 않은 경우는 경고음으로?
    # L.count(1)은 프레임 마다 검사를 했을 때, No_Helmet이라고 판단하는 변수가 몇 이상일 때 경보음을 울릴 것인가 경계값을 정하는 변수 
    if(frame_id % 10==0 and L.count(1)>=3): # 우선 test용으로 10초에 한번씩 노헬멧이 3번이상 인식되면 경보음
        print("헬멧 미착용")
        sound()

    cv2.imshow("Helmet Detection", frame)
#     cv2.moveWindow('Helmet Detection', 200, 200)              # 창 위치 변경
#     cv2.resizeWindow('Helmet Detection', 416, 416)           # 창 크기 변경 (변경 됨))
    key = cv2.waitKey(1)
    if key == ord('q'):  # q 두번 누르면 창 닫힘
        break

cap.release()
cv2.waitKey()
cv2.destroyAllWindows()
# 핸드폰 카메라와 연결하고 yolo 실행확인했으나 속도가 너무 느려서 영상->사진으로 바꿔야 할듯
```
